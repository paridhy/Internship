```python
#Creating empty list to store product urls 
product_url=[]

#Iterating 3 pages
start=0
end=3
for i in range(start,end):
    print('Scraping',i+1)
    #Scraping product urls
    for p in driver.find_elements(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
        product_url.append(p.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[29]/div/div/span/a[3]")
   
```

```python
len(product_url)
```

```python
#making empty lists to store specific information
brand=[]
name=[]
price=[]
exchange=[]
delivery=[]
availability=[]

#iterating in product urls to open product pages
for i in product_url:
    driver.get(i)
    sleep(3)

#scraping brand  
    try:
        p=driver.find_element(By.XPATH,"(//span[@class='a-size-base po-break-word'])[1]")
        brand.append(p.text)
    except NoSuchElementException :
        brand.append('-')
        
#scraping product name 
    try:
        q=driver.find_element(By.XPATH,"(//span[@class='a-size-base po-break-word'])[2]")
        name.append(q.text)
    except NoSuchElementException:
        name.append('-')
            
#scraping price
   
    try:
        s=driver.find_element(By.XPATH,"(//span[@class='a-price-whole'])[1]")
        price.append(s.text)
    except NoSuchElementException :
        price.append('-')

#scraping exchange        
    try:
        t=driver.find_element(By.XPATH,"(//span[@class='a-size-small a-color-link a-text-normal'])[1]")
        exchange.append(t.text)
    except NoSuchElementException :
        exchange.append('-')

#scraping delivery        
    try:
        u=driver.find_element(By.XPATH,"(//span[@class='a-text-bold'])[1]")
        delivery.append(u.text)
    except NoSuchElementException :
        delivery.append('-')
        
#scraping price for first 10 items        
    try:
        v=driver.find_element(By.XPATH,"(//span[@class='a-size-medium a-color-success'])[1]")
        availability.append(v.text)
    except NoSuchElementException :
        availability.append('-')
        

driver.quit()
```

```python
df=pd.DataFrame({'Brand':brand,'Product Name':name,'Price':price,'Replacement':exchange,'Estimated Delivery':delivery,'Availability':availability,'Product URL':product_url})
df
```

```python
# Save DataFrame to CSV
df.to_csv('earphonedata.csv')
```

```python
rank=[]
bl_name=[]
net_worth=[]
age=[]
city=[]
source=[]
ind=[]
delay=5
sleep(5)

# giving a delay till element is located
WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/div[2]/div[3]/div[1]/div[2]/div/h2")))

#scraping ranks
a=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][1]/div")
try:
    for ranks in a:
        rank.append(ranks.text)
except Exception:
    print('not found')
    rank.append('-')

#scraping names
try:
    b=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][2]/div")
    for i in b:
        bl_name.append(i.text)
except Exception:
    print('not found')
    bl_name.append('-')

#scraping net-worth
try:
    c=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][3]/div")
    for i in c:
        net_worth.append(i.text)
except Exception:
    print('not found')
    net_worth.append('-')

#scraping age
try:
    d=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][4]/div")
    for i in d:
        age.append(i.text)
except Exception:
    print('not found')
    age.append('-')  

#scraping citizenship by arithmetic progression
try:
    start = 5
    # nth term = start term + (no of terms * step size)
    end = 5 + (199 * 7)
    step = 7
    for i in range(start, end + 1, step):
        e=driver.find_elements(By.XPATH,f"(//div[@class='TableRow_cell__db-hv Table_cell__houv9'])[{i}]")
        for element in e:
            city.append(element.text)
except Exception:
    print('not found')
    city.append('-')

#scraping source
try:
    f=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][6]/div")
    for i in f:
        source.append(i.text)
except Exception:
    print('not found')
    source.append('-')

#scraping industry
try:
    g=driver.find_elements(By.XPATH,"//div[@class='TableRow_cell__db-hv Table_cell__houv9'][7]/div")
    for i in g:
        ind.append(i.text)
except Exception:
    print('not found')
    ind.append('-')   
```

```python
hostel_url=[]

#getting hostel urls  
try:
    for urls in driver.find_elements(By.XPATH,"//a[@class='property-card-container horizontal']"):
        hostel_url.append(urls.get_attribute('href'))   

except Exception:
    print('done')
```

```python
len(hostel_url)
```

```python
hostel_name=[]
dist=[]
ratings=[]
overall=[]
total_review=[]

try:
    #scraping hostel names
    for name in driver.find_elements(By.XPATH,"//div[@class='property-name']/span"):
            hostel_name.append(name.text)
        
    #scraping distance from city centre
    try:
        for dis_ in driver.find_elements(By.XPATH,"//span[@class='distance-description']"):
            dist.append(dis_.text) 
    except NoSuchElementException:
            dist.append('-')

    #scraping ratings
    try:        
        for r in driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div/span[@class='number']"):
            ratings.append(r.text)
    except NoSuchElementException:
        ratings.append('-')

    #scrapping overall review
    try:    
        for ore in driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div/span[@class='keyword']"):
            overall.append(ore.text) 
    except NoSuchElementException:
        overall.append('-')

    #scraping total reviews
    try:    
        for tr in driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div/span[@class='left-margin']"):
            total_review.append(tr.text) 
    except NoSuchElementException:
        total_review.append('-')

except Exception:
    print('done')
```

```python
facility=[]
pdes=[]
delay=2

#iterating in every hostel url
for i in hostel_url:
    driver.get(i)
    
    #getting specific info of every product
    try:
        pd_=driver.find_element(By.XPATH,"//div[@class='content collapse-content']")
        pdes.append(pd_.text)
    except Exception :
        print('No Property description Available')
        pdes.append('-')

    try:
        loc_fac=driver.find_element(By.XPATH,"(//div[@class='content'])[3]")
        loc_fac.click()
        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH,"(//ul[@class='facilities'])[1]")))
        
        #creating another list to store facilities
        items=[]
        for i in driver.find_elements(By.XPATH,"(//ul[@class='facilities'])[1]/li"):
            items.append(i.text)
        #appending items into facility    
        facility.append(items)    
        
    except Exception :
        print('No facility Available')
        facility.append('-')
```

```python
df9=pd.DataFrame({'Hostel Name':hostel_name,'Distance from city centre':dist,'Ratings':ratings,'Overall Review':overall,'Total Reviews':total_review,'Facilities':facility,'Property Description':pdes})
df9
```
