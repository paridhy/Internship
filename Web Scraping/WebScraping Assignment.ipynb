```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#QUESTION 1
#sending request to the URL and storing response in the page variable
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
#extracting page HTML content using BeautifulSoup and storing it in soup variable
soup=BeautifulSoup(page.content)
```

```python
#creating empty list to store all the headers
headers=[]
for i in soup.find_all('h1'):
    
    #append the list for all h1 tags
    headers.append(i.text)
    
for i in soup.find_all('h2',class_='mp-h2'):
    
    #append the list for all h2 tags
    headers.append(i.text)    
  
```

```python
#create a dataframe to display headers
df=pd.DataFrame({'headers': headers})
df
```

```python
#QUESTION 2

#sending request to the URL and storing response in the page_2 variable
page_2=requests.get('https://presidentofindia.nic.in/index.php/former-presidents')
page_2
```

```python
#extracting page HTML content using BeautifulSoup and storing it in presidents variable
presidents= BeautifulSoup(page_2.content)
```

```python
#find all <div> elements with class 'views-row' and storing it in div variable
div=presidents.find_all('div', class_="views-row")
```

```python
#create empty lists for name and chronology
name=[]
chronology=[]

#run for loop on div elements
for i in div:
    
    #append name list with h3 elements in text format
    name.append(i.find('h3').text.strip())  #strip() to eliminate any trailing spaces
    
    #append chronology list with h5 elements in text format 
    chronology.append(i.find('h5').text.strip()) #strip() to eliminate any trailing spaces
    
    
    
```

```python
#create empty list to store images
images=[]

#find all elements with img tag from presidents variable
for i in presidents.find_all('img',loading='lazy'):
    
    #append with src format
    images.append(i['src'])
 
    
```

```python
#Create a dataframe to display results
df_2=pd.DataFrame({'Name':name,'Order':chronology,'Image':images})
df_2
```

```python
#QUESTION 5

#sending request to the URL and storing response in the page_4 variable
page_4=requests.get('https://www.cnbc.com/world/?region=world')
page_4
```

```python
#extracting page HTML content using BeautifulSoup and storing it in news variable
news=BeautifulSoup(page_4.content)
```

```python
#create empty list to store headlines
headline=[]

for i in news.find_all('a',class_='LatestNews-headline'):
    
    #append headline list in text format
    headline.append(i.text)
    
#create empty list to store time    
time=[]

for i in news.find_all('time',class_='LatestNews-timestamp'):
    
    #append time list in text format
    time.append(i.text)
    
#create empty list to store link    
link=[]
for i in news.find_all('a',class_='LatestNews-headline'):
    
    #append link list in href format
    link.append(i['href'])
```

```python
#create dataframe to display headline,time and link
df_5=pd.DataFrame({'Headline': headline, 'Time of news': time,'Link':link})
df_5
```

```python
#QUESTION 6

#sending request to the URL and storing response in the page_6 variable
page_6=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page_6
```

```python
#extracting page HTML content using BeautifulSoup and storing it in papers variable
papers=BeautifulSoup(page_6.content)
```

```python
#ceating empty lists to store title,authors,date and url of papers
title=[]
authors=[]
date=[]
url=[]

for i in papers.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    
    #append title list in text format
    title.append(i.text)
    
for i in papers.find_all('span',class_='sc-1w3fpd7-0 dnCnAO'):
    
    #append authors list in text format
    authors.append(i.text)  
    
for i in papers.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    
    #append date list in text format
    date.append(i.text) 
    
for i in papers.find_all('a',class_='sc-5smygv-0 fIXTHm'):
    
    #append url list in href format
    url.append(i['href'])    
```

```python
#create a dataframe to display paper title, authors, published date and paper url
df_6=pd.DataFrame({'Paper Title':title,'Authors':authors,'Published Date': date,'Paper URL':url})
df_6
```

```python
#QUESTION 7
#sending request to the URL and storing response in the page_7 variable
page_7=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page_7
```

```python
#extracting page HTML content using BeautifulSoup and storing it in restaurant variable
restaurant=BeautifulSoup(page_7.content)
```

```python
#find all <div> elements with class 'restnt-main-wrap clearfix' and storing it in section variable
section=restaurant.find_all('div',class_='restnt-main-wrap clearfix')
```

```python
#create empty lists to store name, location,rating,image
name=[]
location=[]
rating=[]
image=[]

#apply for loop on section variable
for i in section:
    
    #append name list with elements with a tag and class in text format, strip() eliminates spaces
    name.append(i.find('a',class_="restnt-name ellipsis").text.strip())
    
    #append location list with elements with div tag and class in text format, strip() eliminates spaces
    location.append(i.find('div',class_="restnt-loc ellipsis").text.strip())
    
    #append rating list with elements with div tag and class in text format, strip() eliminates spaces
    rating.append(i.find('div',class_="restnt-rating rating-4").text.strip())
    
    #creating a variable url to find and store img tags with class 'no-img'
    url=i.find('img',class_='no-img')
    
    #append image list with url elements in data-src format
    image.append(url['data-src'])
```

```python
#creating a variable to store all elements with div tag and class 'detail-info'
div_1=restaurant.find_all('div',class_="detail-info")
div_1

#creating a list to store seperate items
cuisine_items = []

for i in div_1:
    # create a list for each div element
    items_in_div = []
    
    #create a new variable to find and store all <a> tag elements in div_1
    new = i.find_all('a')
    for j in new:
        
        #append the list for j elements in text format
        items_in_div.append(j.text)
    
    # Append the list for each div element to the main list
    cuisine_items.append(items_in_div)
cuisine_items
```

```python
#create a dataframe to display restaurant name, cuisine,location,ratings and image url
df_7=pd.DataFrame({'Restaurant Name': name,'Cuisine': cuisine_items,'Location':location,'Ratings': rating,'Image URL': image})
df_7
```
