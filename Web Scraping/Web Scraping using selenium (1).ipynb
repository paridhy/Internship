```python
#creating dataframe and storing the data
df=pd.DataFrame({'Job Title':job_title, 'Job Location':job_location,'Company Name':company_name,'Experience Required':experience })
df
```

```python
#making empty lists to store specific information
brand_name=[]
price_name=[]
descs_name=[]
offs_name=[]
count=100


#for page 1
for i in range(1,2):
    
    #scraping brand name till 100 count
    for w in driver.find_elements(By.XPATH,"//div[@class='_2WkVRV']"):
        brand_name.append(w.text) 
        
    #scraping prices till 100 count   
    for j in driver.find_elements(By.XPATH,"//div[@class='_30jeq3']"):
        price_name.append(j.text)
        
    #scraping description till 100 count    
    for k in driver.find_elements(By.CLASS_NAME,'IRpwTa'):
        descs_name.append(k.text)
        
    #scraping offers till 100 count    
    for k in driver.find_elements(By.CLASS_NAME,'_3Ay6Sb'):
        offs_name.append(k.text)   

        
    #next page button for page 1    
    next_button=driver.find_element(By.XPATH,"/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]/span")
    next_button.click()
    #sleep for 3 secs
    sleep(3)


#for page 2 and beyond    
for i in range(2,4):
    
    
    #scraping brand name till 100 count
    for w in driver.find_elements(By.XPATH,"//div[@class='_2WkVRV']"):
        if len(brand_name)<count:
            brand_name.append(w.text)
            
    #scraping prices till 100 count        
    for j in driver.find_elements(By.XPATH,"//div[@class='_30jeq3']"):
        if len(price_name)<count:
            price_name.append(j.text)
    
    #scraping description till 100 count       
    for k in driver.find_elements(By.CLASS_NAME,'IRpwTa'):
        if len(descs_name)<count: 
            descs_name.append(k.text)
            
    #scraping offers till 100 count        
    for k in driver.find_elements(By.CLASS_NAME,'_3Ay6Sb'):
        if len(offs_name)<count: 
            offs_name.append(k.text)
            
    #next page button for page 2 and beyond          
    next__=driver.find_element(By.XPATH,"/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[12]/span")
    next__.click()    
    #sleep for 2 secs
    sleep(2)
    
       
```

```python
carname=[]
price=[]
count=50

#scraping car names
for w in driver.find_elements(By.XPATH,"//h3[@class='subheader']"):
    if len(carname)<count:
        carname.append(w.text) 
        
#scraping price    
for j in driver.find_elements(By.XPATH,"//strong[contains(text(),'$')]"):
    price.append(j.text.replace('Price:',''))
        
```

```python
#creating dataframe and storing the data
df_10=pd.DataFrame({'Car Name':carname,'Price':price})
df_10
```
